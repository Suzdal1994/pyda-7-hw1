## Задание 10. проверить несколько гипотез для данных, чтобы понять почему сотрудники покидают компанию

**Проверим гипотезу, что из компании увольняются люди, у которых не было повышения в течение последних 5 лет.** 

```python
test1 = data[data['promotion_last_5years']==0].left
test2 = data[data['promotion_last_5years']==1].left

descriptives, results = rp.ttest(test1, test2)

print(descriptives)
print('_______________________________________')
print(results)
```

Difference < 0 p value = 1.0000, то отличие людей с показателем promotion_last_5years = 0 не значительное. Значит, люди с показателем promotion_last_5years=1 не увольняются чаще, чем остальные 

**Проверим гипотезу, что из компании увольняются люди, при time_spend_company менее 5** 

```python
test3 = data[data['time_spend_company']<5].left
test4 = data[data['time_spend_company']>=5].left

descriptives, results = rp.ttest(test3, test4)

print(descriptives)
print('_______________________________________')
print(results)
```

Difference > 0 p value = 1.0000, то отличие людей с показателем time_spend_company < 5 значительное. Значит, люди с показателем time_spend_company < 5 увольняются чаще, чем остальные
